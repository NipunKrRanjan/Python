<h3> GridSearchCV for 1 file </h3>

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
df=pd.read_csv(r'IRIS.csv')
rfc=RandomForestClassifier()
clf=GridSearchCV(rfc,{
    'n_estimators':[40,50,100],
    'max_depth':[5,7],
    #'max_features':['sqrt', 'log2']
},cv=5,return_train_score=False)
print(df.info())
clf.fit(df.drop('species',axis='columns'),df['species'])
print("Best Parameter",clf.best_params_)
print(f"Best Score={(clf.best_score_*100):.2f}")
results = pd.DataFrame(clf.cv_results_)
#print(results.info())
res = results[['mean_test_score', 'params']]
print(res)
```

<h3>GridSearch CV for various models </h3>

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LinearRegression
rf=RandomForestClassifier()
sv=SVC()
lr=LinearRegression()
model_prams={
    'svm':{
            'model':SVC(),
            'params':{
                    'kernel':['linear', 'poly', 'rbf', 'sigmoid'],
                    'C':[1,10,20]
                    }
        }, #in the format of key:value 
    'RandomForest':{
        'model':RandomForestClassifier(),
        'params':{
            'n_estimators':[40,50,100],
            'max_depth':[5,7],
        }
    }
}
scores=[]
df=pd.read_csv(r'IRIS.csv') 
for model_name,mp in model_prams.items(): #accessing using the format key,value 
    clf=GridSearchCV(mp['model'],mp['params'],cv=5,return_train_score=False)
    clf.fit(df.drop('species',axis='columns'),df['species'])
    scores.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })
res=pd.DataFrame(scores,columns=['model','best_score','best_params'])
print(res)
```
