<h3> Plotting using MatplotLib </h3>

```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
X,Y=make_classification(n_classes=4,n_samples=1000,n_features=5,n_informative=3)
plt.plot(X[:,0],Y)
plt.xlabel("X ")
plt.ylabel("Target")
plt.show()
```

<h3>Using Seaborn </h3>

```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
X,Y=make_classification(n_classes=4,n_samples=1000,n_features=5,n_informative=3)
df=pd.DataFrame(X,columns=['F1','F2','F3','F4','F5'])
df['Y']=Y
print(df)
sns.pairplot(data=df,hue='Y')#color="#1A0C82")
```
