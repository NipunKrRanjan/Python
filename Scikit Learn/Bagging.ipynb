```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import accuracy_score,classification_report
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
df=pd.read_csv(r"C:\Users\KIIT0001\Desktop\Python\CSV\diabetes.csv")
lr=LogisticRegression()
baggged=BaggingClassifier(
    estimator=DecisionTreeClassifier(),
    n_estimators=100,
    max_samples=0.8,
    bootstrap=True,
    oob_score=True,
)
X=StandardScaler().fit_transform(df.drop('Outcome',axis='columns'))
Y=df.Outcome
X_train, X_test, y_train, y_test =train_test_split(X,Y,test_size=0.3,shuffle=True,stratify=Y,random_state=42)
baggged.fit(X_train,y_train)
lr.fit(X_train,y_train)
pred_bg=baggged.predict(X_test)
pred_lr=lr.predict(X_test)
print(f"Accuracy of Bagged Decision Tree={accuracy_score(y_test,pred_bg)*100:.3f}%")
print("Accuracy Metrices for Decision Tree:\n",classification_report(y_test,pred_bg))
print(f"Accuracy of Logistic Regression={accuracy_score(y_test,pred_lr)*100:.3f}%")
print("Accuracy Metrices for Logistic Regression:\n",classification_report(y_test,pred_lr))
```
