<h3> Binary Classification </h3>

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score,confusion_matrix
sc=pd.read_csv(r'Social_Network_Ads.csv')
#print(sc.info())

#Label Encoding Gender
gen=sc['Gender']
le=LabelEncoder()
gen_ec=le.fit_transform(gen)
sc['Encoded_Gender']=gen_ec
#print(sc.info())

#Dropping Gender Column
sc=sc.drop(['Gender','User ID'],axis='columns')
#print(sc)
Y=sc['Purchased']
X=sc.drop('Purchased',axis='columns')
model=LogisticRegression()
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.8,random_state=42)
#sns.pairplot(data=sc)
model.fit(X_train,y_train)
pred=model.predict(X_test)
Predicted_Value=np.where(pred==1,'PURCHASED','NOT PURCHASED').reshape(-1,1)
#print(Predicted_Value)
print(accuracy_score(pred,y_test))
cm=confusion_matrix(y_test,pred)
print(cm)
sns.heatmap(cm,annot=True)
plt.show()
```
