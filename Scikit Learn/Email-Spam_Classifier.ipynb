<h3> Without Pipeline </h3>

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
info=pd.read_csv(r'C:\Users\KIIT0001\Desktop\Python\CSV\email.csv')
print(info.info())
info['Category']=info['Category'].apply(lambda x:'Not Spam' if x=='ham' else 'spam')
cv=CountVectorizer()
transformed_cv=cv.fit_transform(info['Message'])
X_train, X_test, y_train, y_test =train_test_split(transformed_cv,info['Category'],test_size=0.25,shuffle=True)
mnb=MultinomialNB()
mnb.fit(X_train,y_train)
pred_sp=mnb.predict(X_test)
acc=accuracy_score(y_test,pred_sp)*100
print(f"Accuracy={acc:.2f}")
```

<h3> Using Pipeline </h3

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
clf=Pipeline([
    ('vt',CountVectorizer()),
    ('nb',MultinomialNB())
])
info=pd.read_csv(r'C:\Users\KIIT0001\Desktop\Python\CSV\email.csv')
test=pd.read_csv(r'C:\Users\KIIT0001\Desktop\Python\CSV\spam.csv')
info['Category']=info['Category'].apply(lambda x:'Not Spam' if x=='ham' else 'spam')
test['Category']=test['Category'].apply(lambda x:'Not Spam' if x=='ham' else 'spam')
X_train, X_test, y_train, y_test =train_test_split(info['Message'],info['Category'],test_size=0.25,shuffle=True)
clf.fit(X_train,y_train)
#test=pd.read_csv(r'C:\Users\KIIT0001\Desktop\Python\CSV\spam.csv')
print(test.info())
pred=clf.predict(test['Message'])
print("Testing data=",info['Category'].head(20))
print("Predicted Output=",pred[:20])

print(f"Accuracy Score={accuracy_score(test['Category'],pred)*100:.2f}")
```
