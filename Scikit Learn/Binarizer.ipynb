```python
from sklearn.preprocessing import Binarizer
import pandas as pd
import numpy as np
df=pd.read_csv(r'C:\Users\KIIT0001\Desktop\Python Practice\CSV\Simple 100 Student Marks.csv')
bina=Binarizer(threshold=40)
bin1=bina.transform(df[['maths_marks']]).flatten()
df['Maths_result']=np.where(bin1==1,'PASS','FAIL')
# This will overwrite your original file
df.to_csv(r'C:\Users\KIIT0001\Desktop\Python Practice\CSV\Simple 100 Student Marks(updated).csv', index=False)
print(df)
```
