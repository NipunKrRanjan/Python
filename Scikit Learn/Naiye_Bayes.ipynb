<h3>Using Titanic Dataset </h3>

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
df=pd.read_csv(r'C:\Users\KIIT0001\Desktop\Python\CSV\Titanic.csv')
print(df.info())
kaam_ka_data=df[['Sex','Pclass','Age','Fare']]
output=df[['Survived']]
nb=GaussianNB()
l_enc=LabelEncoder()
imp=SimpleImputer()
#print(kaam_ka_data)
#print(kaam_ka_data)
missing=kaam_ka_data.columns[kaam_ka_data.isna().any()]
kaam_ka_data['Age']=imp.fit_transform(kaam_ka_data[['Age']])
kaam_ka_data['Fare']=imp.fit_transform(kaam_ka_data[['Fare']])
print(kaam_ka_data.columns[kaam_ka_data.isna().any()])
kaam_ka_data['Sex']=l_enc.fit_transform(kaam_ka_data['Sex'])
print(kaam_ka_data)
X_train, X_test, y_train, y_test =train_test_split(kaam_ka_data,output,test_size=0.2,shuffle=True)
nb.fit(X_train,y_train)
pred=nb.predict(X_test)
acc_sc=accuracy_score(y_test,pred)*100
print(f"Accuracy Score={acc_sc:.2f}")
```
