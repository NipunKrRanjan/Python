<h3>Saving Model using pickle </h3>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
from sklearn.model_selection import train_test_split
import pickle
#import Math
dt=pd.read_csv(r'Salary_dataset.csv')
print(dt.info())
dt_train=pd.DataFrame(dt.drop('Salary',axis=1))
X=dt['YearsExperience']
X_train,X_test,Y_Train,Y_Test=train_test_split(dt['YearsExperience'],dt['Salary'],test_size=0.2,shuffle=True,random_state=42)
llf=linear_model.LinearRegression()
X_train=np.array(X_train).reshape(-1,1)
X_test=np.array(X_test).reshape(-1,1)
llf.fit(X_train,Y_Train)
llf.score(X_test,Y_Test)
p_values=llf.predict(X_test).reshape(-1,1)
with open('LinearRegression','wb')as f:
    pickle.dump(llf,f)
with open('LinearRegression','rb') as f:
    mod=pickle.load(f)
print(mod)
```

<h3>Using JobLib </h3>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
from sklearn.model_selection import train_test_split
import joblib
#import Math
dt=pd.read_csv(r'Salary_dataset.csv')
print(dt.info())
dt_train=pd.DataFrame(dt.drop('Salary',axis=1))
X=dt['YearsExperience']
X_train,X_test,Y_Train,Y_Test=train_test_split(dt['YearsExperience'],dt['Salary'],test_size=0.2,shuffle=True,random_state=42)
llf=linear_model.LinearRegression()
X_train=np.array(X_train).reshape(-1,1)
X_test=np.array(X_test).reshape(-1,1)
llf.fit(X_train,Y_Train)
llf.score(X_test,Y_Test)
p_values=llf.predict(X_test).reshape(-1,1)
joblib.dump(llf,'LinearRegression_Joblib')
```
