<h3>Plotting using MatplotLib </h3>

```python
from sklearn.datasets import make_regression
import pandas as pd
# 1. Generate the data
X,Y = make_regression(n_samples=100, n_features=3, noise=10, random_state=42)
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
plt.grid()
plt.plot(X,Y,marker='.')
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
print(np.min(Y))
```

<h3> Plotting using Seaborn </h3>

```python
from sklearn.datasets import make_regression
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Generate data
X, y = make_regression(n_samples=100, n_features=3, noise=10, random_state=42)

# 2. Convert to DataFrame for Seaborn
df = pd.DataFrame(X, columns=['Feature_1', 'Feature_2', 'Feature_3'])
df['Target'] = y

# 3. Plot using Seaborn
sns.scatterplot(data=df, x='Feature_1', y='Target', hue='Target', palette='viridis')
plt.title("Seaborn Scatter Plot of Regression Data")
plt.show()
```
