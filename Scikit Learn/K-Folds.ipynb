<h3> Implementing K-folds </h3>

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold
from sklearn.datasets import load_iris
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(max_iter=200)
rf=RandomForestClassifier(n_estimators=50,max_features='sqrt')
svm=SVC()
info=load_iris()
#print(dir(info))
x=info.data
y=info.target
#print(len(x))
result=info.target_names[y]
print(len(x),len(result))
folds=10
#Defining the function to return the accuracy score
def get_score(model,X_train,X_test,Y_train,Y_test):
    model.fit(X_train,Y_train)
    return model.score(X_test,Y_test)

kf=KFold(
    n_splits=folds,
    shuffle=True,
    random_state=42
    )
score_lr=[]
score_rf=[]
score_svm=[]
for train_index,test_index in kf.split(x):
    X_train,X_test,Y_train,Y_test=x[train_index],x[test_index],result[train_index],result[test_index]
    score_lr.append(get_score(lr,X_train,X_test,Y_train,Y_test))
    score_rf.append(get_score(rf,X_train,X_test,Y_train,Y_test))
    score_svm.append(get_score(svm,X_train,X_test,Y_train,Y_test))
# Note the 'f' at the start and the removal of the comma before \n
print(f"Average score of Logistic Regression = {np.mean(score_lr)*100:.2f}%"
      f"\nAverage score of Random Forest       = {np.mean(score_rf)*100:.2f}%"
      f"\nAverage score of SVM                 = {np.mean(score_svm)*100:.2f}%")
```

<h3> Implementing Cross Val Score</h3>

```python
import numpy as np
from sklearn.model_selection import KFold, cross_val_score
from sklearn.datasets import load_iris
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

# 1. Load Data
info = load_iris()
X = info.data
y = info.target  # Using integers (0, 1, 2) is best practice for training

# 2. Define Models
lr = LogisticRegression(max_iter=200)
rf = RandomForestClassifier(n_estimators=50, max_features='sqrt')
svm = SVC()

# 3. Define the K-Fold Strategy
# Using a KFold object allows us to SHUFFLE the data, which is vital for Iris
folds = 10
kf = KFold(n_splits=folds, shuffle=True, random_state=42)

# 4. Run Cross-Validation for each model
# cross_val_score returns an array of 10 scores (one for each fold)
score_lr = cross_val_score(lr, X, y, cv=kf)
score_rf = cross_val_score(rf, X, y, cv=kf)
score_svm = cross_val_score(svm, X, y, cv=kf)

# 5. Calculate and Print Averages
print("--- Cross-Validation Results (Accuracy) ---")
print(f"Logistic Regression: {np.mean(score_lr)*100:.2f}%")
print(f"Random Forest:       {np.mean(score_rf)*100:.2f}%")
print(f"SVM:                 {np.mean(score_svm)*100:.2f}%")
```
