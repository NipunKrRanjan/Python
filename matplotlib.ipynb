```python
from matplotlib import pyplot as plt
import pandas as pd
from matplotlib import style
import numpy as np
style.use('seaborn-v0_8')
df1=pd.read_csv(r'CSV\mdb_movies.csv')
x=df1['runtime'].iloc[0:25]
y=df1['revenue'].iloc[0:25]
plt.bar(x,y)
plt.title("Runtime Vs Revenue")
plt.ylabel('Revenue')
plt.xlabel('Runtime')
plt.show()
```
