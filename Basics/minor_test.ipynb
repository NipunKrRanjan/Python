<h1> Testing how to plot a graph</h1>

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick #used for handling ticks and labesl
import matplotlib.dates as mdates #can be used for conversion into date time such that it could be plotted
import seaborn as sns

df = pd.read_csv(r'JSON\CSV\BRTI Historical Data.csv')
print(df.info())

#Convert the originally string dates to python date-time object 
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')


#converts the originally string format into float type objects for treatment as numbers
df['Price'] = df['Price'].str.replace(',', '').astype(float)

sns.set_theme(style="whitegrid", rc={"grid.linestyle": "--"})
plt.figure(figsize=(14, 7), dpi=100)

#Fills the gap covered by the graph
plt.fill_between(df['Date'], df['Price'], color="#3498db4a", alpha=.1)
ax = sns.lineplot(x='Date', y='Price', data=df, color="#2980b9", linewidth=2.5)

#Decides the template of how to display numbers on y-axis
ax.yaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.0f}'))

#Sets the starting and ending point of the graph rather than zero
ax.set_ylim(df['Price'].min() * 0.95, df['Price'].max() * 1.05)

ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))

plt.title('BRTI Stock Price Trend Analysis', fontsize=18, fontweight='bold', pad=20)
plt.xlabel('Timeline', fontsize=12)
plt.ylabel('Price (INR)', fontsize=12, labelpad=15)

#plt.tight_layout()
plt.show()
```
