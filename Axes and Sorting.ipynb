```python
import numpy as np
arr1=np.array([[1,2,3],[4,5,6]])
print(arr1.sum(axis=0)) # Addition of all elements column wise
print(arr1.sum(axis=1)) # Addition of all elements row wise
print('\n')
arr2=np.array([[7,8,9],[2,8,0]])
print("Dot Product of 2 metrices:\n",np.dot(arr1,arr2.transpose())) 
print('\n')
print("Cross Product of 2 matrices:\n",np.cross(arr1,arr2))
```
