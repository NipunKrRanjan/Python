```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load data and select the first 20 students
var = pd.read_csv(r"CSV\Simple 100 Student Marks.csv").iloc[0:20]

# 1. Set the figure size (Width=12, Height=6 is usually good for 20 names)
plt.figure(figsize=(12, 6))

# 2. Create the plot 
# Note: Ensure the column name (e.g., "name" or "student_name") matches your CSV
sns.stripplot(x="student_name",y="maths_marks", data=var)

# 3. Rotate the x-axis labels so they don't overlap
plt.xticks(rotation=45, ha='right')

# 4. Add some padding so names aren't cut off at the bottom
plt.tight_layout()

plt.show()
#sns.countplot(x="sex",data=var,hue="species",palette="icefire")
```
