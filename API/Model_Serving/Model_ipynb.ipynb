```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,recall_score,confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn import tree
import pickle
iris=load_iris()
print(dir(iris))
dt=pd.DataFrame(iris.data,columns=iris.feature_names)
#print(dt)
dt['Result']=iris.target_names[iris.target]
Y=dt['Result']
X=iris.data
#print(dt)
rfc=RandomForestClassifier(n_estimators=300,max_depth=10,max_features='sqrt',min_samples_split=2)
X_train, X_test, y_train, y_test=train_test_split(X,Y,test_size=0.2,shuffle=True,stratify=Y,random_state=42)
rfc.fit(X_train,y_train)
pred=rfc.predict(X_test)
print("Accuracy score=",accuracy_score(y_test,pred))
#print("Recall score=",recall_score(y_test,pred,labels='average'))
# Access the first tree (index 0)
plt.figure(figsize=(100,150))
tree.plot_tree(rfc.estimators_[0], 
               feature_names=iris.feature_names, 
               class_names=iris.target_names, 
               filled=True, 
               rounded=True)
plt.show()
with open('Model','wb') as f:
    pickle.dump(rfc,f)
```
